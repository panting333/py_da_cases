```python
data=pd.read_csv('train.csv')
```

```python
data.head()
```

```python
data.info()
```

```python
print(data.isnull().sum())#Age列存在177个空值，Cabin列存在687个空值，Embarked列存在两个空值
```

```python
#第一步：数据清洗（缺失值、异常值、重复值、无关值等）
data.drop(['PassengerId','Name','Ticket','Cabin'],axis=1,inplace=True)#将不会影响到存活率的无关要素列drop掉
```

```python
data.head()
```

```python
data.Age.fillna(data.Age.mean(),inplace=True)#将Age列的缺失值用平均年龄填充
```

```python
data.isnull().sum()
```

```python
data.dropna(subset=['Embarked'],how='any',axis=0,inplace=True)#将Embarked缺失的行drop掉
```

```python
data.isnull().sum()
```

```python
#第二步：数据规整
##性别转换为数值，便于后续进行分析
data.Sex.unique()
```

```python
data
```

```python
data.loc[data['Sex']=='male','Sex']
```

```python
#将性别转为数值，便于后续进行分析，male为0，female为1
data.loc[data['Sex']=='male','Sex']=0
data.loc[data['Sex']=='female','Sex']=1
```

```python
#将登陆港口转为数值,便于后续分析，S为0，C为1，Q为2
data.Embarked.unique()
```

```python
data.loc[data['Embarked']=='S','Embarked']=0
data.loc[data['Embarked']=='C','Embarked']=1
data.loc[data['Embarked']=='Q','Embarked']=2
```

```python
pd.options.display.max_rows=None#Notebook 的一个cell的显示行数
pd.options.display.max_columns=None#Notebook 的一个cell的显示列数
data.describe()
```

```python
data['Survived'].value_counts()
```

```python
#数据分析
##存活人数分布
plt.rcParams['font.sans-serif']=['SimHei']#解决中文乱码,使用指定的汉字字体类型（此处为黑体）
fig,axes=plt.subplots(2,3,figsize=(8,8))
fig.subplots_adjust(wspace=0.4,hspace=0.3)
data['Survived'].value_counts().sort_values().plot(kind='bar',alpha=0.8,ax=axes[0][0],title='存活人数分布')
axes[0][0].set_xlabel('性别')
axes[0][0].set_ylabel('人数')
```

```python
##不同船舱的人数分布
data['Pclass'].value_counts().sort_values().plot(kind='bar',ax=axes[0][1],alpha=0.8,title='不同船舱的人数分布')
axes[0][1].set_xlabel('船舱')
axes[0][1].set_ylabel('人数')
```

```python
fig
```

```python
#存活和遇难的年龄分布
axes[0][2].scatter(data['Survived'],data['Age'],alpha=0.8)
axes[0][2].set_title('存活和遇难年龄分布')
axes[0][2].set_xlabel('是否存活')
axes[0][2].set_ylabel('年龄')

```

```python
fig
```

```python
#不同港口登陆的人数分布
data['Embarked'].value_counts().sort_values().plot(kind='bar',ax=axes[1][0],alpha=0.8,title='不同港口登陆的人数分布')
axes[1][0].set_xlabel('港口')
axes[1][0].set_ylabel('人数')
```

```python
fig
```

```python
#不同仓位的年龄分布
data['Age'][data['Pclass']==1].plot(kind='kde',label='1',ax=axes[1][1])
data['Age'][data['Pclass']==2].plot(kind='kde',label='2',ax=axes[1][1])
data['Age'][data['Pclass']==3].plot(kind='kde',label='3',ax=axes[1][1])
axes[1][1].set_title('不同舱位的年龄分布')
axes[1][1].set_xlabel('年龄')
axes[1][1].set_ylabel('密度')
axes[1][1].legend(loc='best')
```

```python
fig
```

```python
#不同非直系亲属数量的人数分布
data['SibSp'].value_counts().sort_values().plot(kind='bar',ax=axes[1][2],title='不同非直系亲属数量的人数分布')
axes[1][2].set_xlabel('非直系亲属的数量')
axes[1][2].set_ylabel('人数')

```

```python
fig
```
